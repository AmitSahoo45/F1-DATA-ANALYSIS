```python
import fastf1 as ff1
from fastf1 import plotting
from fastf1 import utils

from matplotlib import pyplot as plt
from matplotlib.pyplot import figure

import numpy as np
import pandas as pd
```

```python
ff1.Cache.enable_cache('../cache')
```

```python
YEAR = 2023
RACE = 'Saudi-Arabia'
SESSION = 'R'

session = ff1.get_session(
    YEAR,
    RACE,
    SESSION
)
```

```python
session.load()
```

```python
MAX, PEREZ = 'VER', 'PER'
```

```python
laps = session.laps
```

```python
maxLaps = laps.pick_driver(MAX)
perezLaps = laps.pick_driver(PEREZ)
```

```python
maxFastest = maxLaps.pick_fastest()
perezFastest = perezLaps.pick_fastest()
```

```python
max_car_data = maxFastest.get_car_data()
perez_car_data = perezFastest.get_car_data()
```

```python
maxTelemetry = maxFastest.get_telemetry()
perezTelemetry = perezFastest.get_telemetry()
```

```python
delta, ref, compare = utils.delta_time(maxFastest, perezFastest)
```

```python
MAXDVR = maxFastest['Team']
PERDVR = '#FF0000'
```

```python
plot_size = [15, 15]
plot_title = f"{session.event.year} {session.event.EventName} - {session.name}"
plot_ratios = [2, 3, 3, 3, 3, 3, 3]
plot_filename = plot_title.replace(" ", "") + ".png"
```

```python
plt.rcParams['figure.figsize'] = plot_size

fig, ax = plt.subplots(7, 1, gridspec_kw={'height_ratios': plot_ratios})

ax[0].title.set_text(plot_title)

ax[0].plot(ref['Distance'], delta, color=ff1.plotting.team_color(MAXDVR))
ax[0].axhline(0)
ax[0].set(ylabel=f'Gap(s) to Perez')

# ----------------------------------------------------------------

ax[1].plot(maxTelemetry['Distance'], maxTelemetry['Speed'], color=ff1.plotting.team_color(MAXDVR))
ax[1].plot(perezTelemetry['Distance'], perezTelemetry['Speed'], color=PERDVR)
ax[1].set(ylabel='Speed')
ax[1].legend(loc="lower right")

ax[2].plot(maxTelemetry['Distance'], maxTelemetry['Throttle'], color=ff1.plotting.team_color(MAXDVR))
ax[2].plot(perezTelemetry['Distance'], perezTelemetry['Throttle'], color=PERDVR)
ax[2].set(ylabel='Throttle')
ax[2].legend(loc="lower right")

ax[3].plot(maxTelemetry['Distance'], maxTelemetry['Brake'], color=ff1.plotting.team_color(MAXDVR))
ax[3].plot(perezTelemetry['Distance'], perezTelemetry['Brake'], color=PERDVR)
ax[3].set(ylabel='Brake')
ax[3].legend(loc="lower right")

ax[4].plot(maxTelemetry['Distance'], maxTelemetry['nGear'], color=ff1.plotting.team_color(MAXDVR))
ax[4].plot(perezTelemetry['Distance'], perezTelemetry['nGear'], color=PERDVR)
ax[4].set(ylabel='Gear')
ax[4].legend(loc="lower right")

ax[5].plot(maxTelemetry['Distance'], maxTelemetry['RPM'], color=ff1.plotting.team_color(MAXDVR))
ax[5].plot(perezTelemetry['Distance'], perezTelemetry['RPM'], color=PERDVR)
ax[5].set(ylabel='RPM')
ax[5].legend(loc="lower right")

ax[6].plot(maxTelemetry['Distance'], maxTelemetry['DRS'], color=ff1.plotting.team_color(MAXDVR))
ax[6].plot(perezTelemetry['Distance'], perezTelemetry['DRS'], color=PERDVR)
ax[6].set(ylabel='DRS')
ax[6].legend(loc="lower right")

plt.savefig(plot_filename, dpi=300, bbox_inches='tight')
plt.show()
```
